
#Immigration to Canada
###Immigrants from 150+ countries between 1980 to 2013

About Dataset
Context

The Main Purpose for uploading this dataset is to analyze the trends and hidden patterns of Immigrants from all over the world to Canada.
Content

This dataset consists of immigrants record from 150+ countries to Canada between 1980 to 2013.


https://www.kaggle.com/datasets/ammaraahmad/immigration-to-canada

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/refs/heads/main/Dados/imigrantes_canada.csv'
dados = pd.read_csv(url)
```

```python
dados.info()
```

```python
dados.head()
```

```python
dados.set_index('País', inplace=True)
```

```python
dados
```

```python
anos= list(map(str, range(1980,2014)))
anos
```

```python
#df=dados['1989'].filter('Região'=='Asia')
df = dados[dados['Região'] == 'Sudeste Asiático']
df
```

```python
brasil = dados.loc['Brasil', anos]
brasil
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
brasil_dict
```

```python
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

```python
plt.figure(figsize=(10,6))
plt.title('Imigração do Brasil para o Canadá')
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.xlabel('Ano')
plt.ylabel('Imigrantes')
plt.xticks(np.arange(1980,2015,step=5).astype(str),rotation=90)
plt.yticks(np.arange(500,3000,step=500))
plt.show()
```

# Desafio: Compação Brasil x Argentina
Chegou a hora de você testar os conhecimentos desenvolvidos durante a aula. Nós criamos um gráfico com a função plt.plot() para analisar as intrigantes tendências de imigração do Brasil para o Canadá, no período de 1980 a 2013. Neste momento temos uma nova demanda: criar um gráfico de linhas comparando os números de imigrantes do Brasil e Argentina para o Canadá, que são os maiores países da América do Sul.

Prepare-se para um mergulho fascinante nas linhas que conectam esses países da América do Sul ao território canadense. Nessa missão, a elaboração desse gráfico pode ser útil para a compreensão das tendências migratórias desses países para o Canadá ao longo do tempo e como elas se comparam entre si. Ao analisar esses fatores, podemos obter uma visão mais abrangente do cenário migratório na América do Sul.

```python
argentina = dados.loc['Argentina', anos]
argentina_dict = {'ano': argentina.index.tolist(), 'imigrantes': argentina.values.tolist()}
dados_argentina = pd.DataFrame(argentina_dict)
dados_argentina
```

```python
brasil_argentina = pd.concat([dados_brasil, dados_argentina], axis=1)
brasil_argentina
```

```python
brasil_argentina.columns = ['ano', 'brasil', 'ano2','argentina']
brasil_argentina = brasil_argentina.drop('ano2', axis=1)
brasil_argentina
```

```python
plt.figure(figsize=(10,6))
plt.title('Imigração para o Canadá')
plt.plot(brasil_argentina['ano'], brasil_argentina['brasil'], label='Brasil')
plt.plot(brasil_argentina['ano'], brasil_argentina['argentina'], label='Argentina')
plt.xlabel('Ano')
plt.ylabel('Imigrantes')
plt.legend()
plt.xticks(np.arange(1980,2015,step=5).astype(str),rotation=90)
plt.yticks(np.arange(500,3000,step=500))
plt.show()
```

Outra forma de fazer a comparação:


```python
df_comparacao = dados.loc[['Brasil', 'Argentina'], anos]
df_comparacao = df_comparacao.T
df_comparacao.head()
```

```python
plt.figure(figsize=(10,6))
plt.title('Imigração para o Canadá')
plt.plot(df_comparacao['Brasil'],label = 'Brasil')
plt.plot(df_comparacao['Argentina'],label ='Argentina')
plt.xlabel('Ano')
plt.ylabel('Imigrantes')
plt.legend()
plt.xticks(np.arange(1980,2015,step=5).astype(str),rotation=90)
plt.yticks(np.arange(500,3000,step=500))
plt.show()
```

```python
fig, axs =plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.5)
fig.suptitle('Imigração para o Canadá 1980 a 2013')
idx = 0;
paises = ['Brasil', 'Argentina', 'Colômbia', 'Peru']
for i in range(2):
  for j in range(2):
    pais = paises[idx]
    axs[i,j].plot(dados.loc[pais,anos])
    axs[i,j].set_title(pais)
    idx += 1

ymin = 0
ymax = 7000
for ax in axs.flat: # ou axs.ravel (menos eficiente=cria uma cópia=np.array, .flat é uma view=iterável)
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set(xlabel='Ano', ylabel='Número de Imigrantes')
  ax.set_ylim(ymin,ymax)
  ax.grid()
  #ax.set_xlabel('Ano')
  #ax.set_ylabel('Número de Imigrantes')

plt.show()
```

```python
def gerar_grafico_paleta(paleta):
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient = 'h', hue='Total', legend=False, palette=paleta)
  ax.set_title('Países com mais imigrantes para o Canadá\n1980 a 2013',loc='left',fontsize=14)
  ax.set_xlabel('Número de Imigrantes', fontsize=14)
  ax.set_ylabel('')
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)
  #plt.grid(linestyle="--")
  plt.show()
```
